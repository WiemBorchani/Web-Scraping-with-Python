```python
#imports
import scrapy 
from scrapy.crawler import CrawlerProcess
from time import sleep
```

```python
class ProductsSpider(scrapy.Spider):
  name= 'products'
  start_urls= ['https://www.wiki.tn/c/ordinateur-pc-portable-13.html?id_category=13&n=620']
  def parse(self, response):
    for product in response.css('div.product-container') :
      sleep(0.5)
      try:
        print(product.css('a.product-name::text').get())
        print(product.css('span.product-price::text').get().replace('DT',''))
        print(product.css('a.product-name').attrib['href'])
        yield
        {
            'name': product.css('a.product-name::text').get() ,
            'price': product.css('span.product-price::text').get().replace('DT','') ,
            'link': product.css('a.product-name').attrib['href'],
        }  
      except Error:
        continue
```

```python
process = CrawlerProcess({
'USER_AGENT': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36',
'FEED_FORMAT': 'json',
'FEED_URI': 'products.json'
})
process.crawl(ProductsSpider)
process.start()
```
