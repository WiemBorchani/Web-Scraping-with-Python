```python
#imports
from selenium import webdriver
from kora.selenium import wd
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.remote.webdriver import WebDriver
import time
from os import error
```

we can access the site with two **methods**

method **1**

```python
#using web driver
Path = "/content/drive/MyDrive/chromedriver.exe"
driver = webdriver.Chrome(Path)
driver.get("https://www.wiki.tn/c/telephonie-tablette-15.html")
print(driver.title)
#close the currebt tab
driver.close()
#quit the entire browser
driver.quit ()
#scrapping site using webdriver
try :
  main =WebDriverWait(driver,10).until(
      EC.presence_of_element_located((BY.ID , "product_list" ))
  ) 
  print (main.text)
except:
  driver.quit()
#return to the previous page
driver.back()
#go back to the original page
driver.forward()
```

method **2**

```python
wd.get("https://www.wiki.tn/c/telephonie-tablette-15.html")
display(wd)
```

-

```python
Search = wd.find_element_by_class_name("search_query")
Search.send_keys("tablette")
Search.send_keys(Keys.RETURN)
print(wd.page_source)
time.sleep (5)
#wd.quit()
```

```python
try: 
  main = wd.find_element_by_id("product_list")
  tittles = main.find_elements_by_class_name("product-container")
  for tittle in tittles:
    header = tittle.find_element_by_class_name("name")
    print(header.text)
except error :
  pass
```

```python
#navigation in a web site
link = wd.find_element_by_partial_link_text("Sécurité réseaux , Electricité & Énergie solaire")
link.click()
```
